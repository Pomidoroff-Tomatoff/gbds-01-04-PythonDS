Получить матрицу `B`, умножив матрицу `A` на матрицу `At` с помощью матричного умножения. Какой размер имеет матрица `B`? Получится ли вычислить обратную матрицу для матрицы `B` и почему?

```python
a = np.arange(12)
A = a.reshape(4, 3)
At = np.transpose(A)
A, At
```

```python
B = A.dot(At)
print(f"Матрица перемножения A и At \n{B}, \nРазмер матрицы: {B.shape}")
```

```python
if B_det := np.linalg.det(B):
    B_inv = np.linalg.inv(B)
    print(f"Детерминант матрицы {B_det =} -- не равен нулю и мы можем получить обратную матрицу: \n{B_inv}")
else:
    print(f"Детерминант матрицы {B_det = } равен {B_det} и, поэтому, мы не можем получить обратную матрицу...")
    print(f"")
```
